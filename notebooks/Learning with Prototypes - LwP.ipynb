### Learning with Prototypes

Remember our previous example of classifying mails to two categories - spam or not spam. This kind of problems is called classificaiton problem. Here there are only two classes. Namely *spam* or *not spam*. This is called a __binary classification__ problem. If there are multiple classes to classify a given set of data or an input, it is called __multiclass classification__. If each input need to be assigned to more than one class, then it is called a __multilabel classification__.

In this current example we will create a sample email which will represent all spam emails we have seen earlier. Any new email which is very close to this representative email or __prototype__ is classified as *spam*. In ML, the word prototype is used to refer to something that is representative or captures the qualities of a class of objects

Let us see how we can use the concept of prototypes in Machine Learning.
